```python
with open('영어가사.txt', 'r') as file:
    text = file.read()
print(text)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wc = WordCloud().generate(text)
```

```python
plt.imshow(wc, interpolation='bilinear')
plt.show()
```

```python
with open('골든.txt', 'r', encoding='utf-8') as file:
    text = file.read()
print(text)
```

```python
wc = WordCloud().generate(text)
plt.imshow(wc)
plt.axis('off')
plt.savefig('골든.png')
plt.show()
```

```python
import numpy as np
from PIL import Image
```

```python
heart_mask = np.array(Image.open('heart_mask.png'))
```

```python
wc = WordCloud(
        width=800, 
        height=800, 
        background_color='white',
        min_font_size=20,
        mask=heart_mask).generate(text)
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.savefig('골든하트.png')
plt.show()
```
