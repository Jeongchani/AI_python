```python
def change_no(no):
    no = no.replace('번', '')
    no = int(no)
    no = f'{no:02d}'
    return no
```

```python
no = change_no('2번')
no
```

```python
df2 = score.copy()
df2.reset_index(inplace=True)
df2['지원번호'] = df2['지원번호'].apply(change_no)
df2.set_index('지원번호', inplace=True)
df2
```

```python
df2.loc['10','이름'] = '홍길동'
df2
```

```python
max=df2.index.max()
new_no=str(int(max)+1)
new_no
```
